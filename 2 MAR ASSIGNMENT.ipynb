```python

x=np.array([0, 1, 2, 3, 4, 5])
y=np.array([0, 100, 200, 300, 400, 500])
plt.subplot(4,5,1)
plt.plot(x,y)

x=np.array([0, 1, 2, 3, 4, 5])
y = np.array([50, 20, 40, 20, 60, 70]) 
plt.subplot(4,5,2)
plt.plot(x,y)


x = np.array([0, 1, 2, 3, 4, 5])
y = np.array([10, 20, 30, 40, 50, 60]) 
plt.subplot(4,5,3)
plt.plot(x,y)

 
x = np.array([0, 1, 2, 3, 4, 5]) 
y = np.array([200, 350, 250, 550, 450, 150])
plt.subplot(4,5,4)
plt.plot(x,y)
```

#Q.5: What is a box plot? Why is it used? Using the following data plot a box plot.
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

ANSWER==A box plot is a statistical representation of the distribution of a variable through its quartiles. The ends of the box represent the lower and upper quartiles, while the median (second quartile) is marked by a line inside the box. For other statistical representations of numerical data, see other statistical charts.

```python
data1 = np.random.normal(100, 10, 200)
data2 = np.random.normal(90, 20, 200)
```

```python
data=[data1,data2]
```

```python
fig = plt.figure(figsize =(10, 9))
 

ax = fig.add_axes([0,0,1,1])

bp = ax.boxplot(data)

plt.show()
```
